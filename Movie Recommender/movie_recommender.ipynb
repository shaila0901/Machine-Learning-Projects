Categorising the movie runtime

```python
def filter(x):
  if(x <= 60):
    return 'ShortFilm'
  elif(x < 120):
    return 'NormalLength'
  else:
    return 'VeryLong'
```

```python
dataset['length'] = dataset['runtime'].apply(filter)
```

```python
dataset = dataset.drop(['runtime' , 'status'] , axis = 1)
```

```python
dataset.head(1)
```

Using the imdb rating metric to give a definitive rating to the movie

```python
v = dataset['vote_count'].astype(int)
r = dataset['vote_average'].astype(int)
C = r.mean()
m = v.quantile(0.75)
```

```python
dataset['rating'] = (v / (v + m)) * r + (m / (v + m)) * C
```

```python
dataset = dataset.drop(['vote_count' , 'vote_average'] , axis = 1)
```

Categorising the hit/flop success of the movie

```python
dataset['profit'] = dataset['revenue'] - dataset['budget']
```

```python
dataset['profit'].describe()
```

```python
def profitapply(x):
  if(x < 0):
    return 'Flop'
  elif(x < 100000):
    return 'Hit'
  else:
    return 'SuperHit'
```

```python
dataset['profit'] = dataset['profit'].apply(profitapply)
```

```python
dataset = dataset.drop(['revenue' , 'budget' , 'popularity'] , axis = 1)
```

Categorising the release date

```python
dataset['release_date'] = dataset['release_date'].astype(str)
```

```python
def datestr(x):
  s = x[0 : 4]
  m = int(s)
  m = m - (m%10)
  s = str(m) + "s"
  return s
```

```python
dataset['year'] = dataset['release_date'].apply(datestr)
```

```python
dataset = dataset.drop('release_date' , axis = 1)
```

```python
import ast
```

```python
def dict_to_l(x):
  l = []
  for i in ((ast.literal_eval(x))[:5]):
    l.append(i['name'].replace(' ' , ''))
  return l
```

Applying the method to dictionary-list tags

```python
dataset['genres'] = dataset['genres'].apply(dict_to_l)
```

```python
dataset['production_companies'] = dataset['production_companies'].apply(dict_to_l)
```

```python
dataset['production_countries'] = dataset['production_countries'].apply(dict_to_l)
```

```python
dataset['keywords'] = dataset['keywords'].apply(dict_to_l)
```

```python
dataset['cast'] = dataset['cast'].apply(dict_to_l)
```

We only want the producer's and director's name from crew, so we shall make a different function for it

```python
def crew_prod(x):
  l = []
  for i in ((ast.literal_eval(x))):
    if i['job'] == 'Executive Producer' or i['job'] == 'Directing' :
      l.append(i['name'].replace(' ' , ''))
  return l
```

```python
dataset['crew'] = dataset['crew'].apply(crew_prod)
```

Creating the tags attribute

```python
def sttolist(x):
  x = str(x)
  x = x.replace(',' , ' ')
  l = []
  l = x.split()
  return l
```

```python
dataset['overview'] = dataset['overview'].apply(sttolist)
```

```python
dataset['tags'] = (dataset['profit'] + " " + dataset["original_language"] + " " + (dataset['length']) + " " + dataset['year'] + " " + (dataset['rating'].astype(str))).str.split()
```

```python
dataset['tags'] = dataset['overview'] + dataset['genres'] + dataset['production_companies'] + dataset['production_countries'] + dataset['keywords'] + dataset['cast'] + dataset['crew']
```

```python
colsdrop = ['overview' , 'genres' , 'production_companies' , 'production_countries' , 'keywords' , 'cast' , 'crew' , 'profit' , 'year' , 'length' , 'rating']
dataset = dataset.drop(colsdrop , axis = 1)
```

# Word Vectorizing

Stemming the words

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem(text):
  y = []
  for i in text:
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
dataset['tags'] = dataset['tags'].apply(stem)
```

Word Vectorizing

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 10000 , stop_words = 'english')
```

```python
vector = cv.fit_transform(dataset['tags'].astype(str)).toarray()
```

```python
vector[0]
```

```python
cv.get_feature_names_out()
```

Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity[0]
```

# Final Recommender System

Recommendation Function

```python
def recomm(xyz):
  i = dataset[dataset['title'] == xyz].index[0]
  dist = similarity[i]
  recomm_list = sorted(list(enumerate(dist)) , reverse = True , key = lambda x:x[1])[1:11]
  for i in recomm_list:
    print(dataset['title'].iloc[i[0]])
```

```python
recomm("Jumanji")
```

```python
recomm('Inception')
```
