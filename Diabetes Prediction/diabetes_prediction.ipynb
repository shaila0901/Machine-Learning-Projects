Zero values

```python
#blood pressure below 90/60 is considered that of a dead person, similarly, bmi<12 is considered for that of starvation.
df.loc[df["BloodPressure"]<60, "BloodPressure"].count()  #121
df.loc[df["BloodPressure"]<60, "BloodPressure"] = 0

df[df["BMI"]<12]['BMI'].count()
df.loc[df["BMI"]<12,"BMI"] = 0
```

```python
#we see some values are 0 but should not be, thus we fill them with nan value
zero_cols = ['Pregnancies' ,	'Glucose' ,	'BloodPressure' ,	'SkinThickness' , 'Insulin'	, 'BMI']
df[zero_cols] = np.where(df[zero_cols] == 0, np.nan, df[zero_cols])
```

```python
#filling in the nan with mean values
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(df)
df = pd.DataFrame(imputer.transform(df))
```

```python
df.columns = ['Pregnancies' ,	'Glucose' ,	'BloodPressure' ,	'SkinThickness' ,	'Insulin' ,	'BMI' ,	'DiabetesPedigreeFunction' ,	'Age' ,	'Outcome']
```

Output and Input dataframes

```python
x = df.iloc[ : , :-1]
y = df["Outcome"]
```

Encoding and Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
# instantiating the scaler
scaler = StandardScaler()

# scaling the continuous featuree
x1 = scaler.fit_transform(x)
```

# Creating the models

Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x1,y, test_size = 0.2, random_state = 2)
```

Importing the models

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
```

Logistic Regression

```python
# instantiating the object
logreg = LogisticRegression()

# fitting the object
logreg.fit(X_train, y_train)

# finding the predicted values
y_pred1 = logreg.predict(X_test)
```

SVM

```python
# instantiating the object and fitting
clf = SVC(kernel='rbf', C=1).fit(X_train,y_train)

# predicting the values
y_pred2 = clf.predict(X_test)
```

Decision tree

```python
# instantiating the object
dt = DecisionTreeClassifier()

# fitting the model
dt.fit(X_train, y_train)

# calculating the predictions
y_pred3 = dt.predict(X_test)
```

Random Forest

```python
# instantiating the object
rf = RandomForestClassifier()

# fitting the model
rf.fit(X_train, y_train)

# calculating the predictions
y_pred4 = rf.predict(X_test)
```

Gradient boost

```python
# instantiate the classifier
gbt = GradientBoostingClassifier(n_estimators = 300,max_depth=1,subsample=0.8,max_features=0.2)

# fitting the model
gbt.fit(X_train,y_train)

# predicting values
y_pred5 = gbt.predict(X_test)
```

KNN

```python
# instantiating the object
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski')

# fitting the object
classifier.fit(X_train, y_train)

# finding the predicted values
y_pred6 = classifier.predict(X_test)
```

Checking accuracies

```python
from sklearn.metrics import accuracy_score
```

```python
print("The test accuracy score of Logistic Regression is ", accuracy_score(y_test, y_pred1))
print("The test accuracy score of SVC is ", accuracy_score(y_test, y_pred2))
print("The test accuracy score of Decision Tree Classifier is ", accuracy_score(y_test, y_pred3))
print("The test accuracy score of Random Forest Classifier is ", accuracy_score(y_test, y_pred4))
print("The test accuracy score of Gradient Boosting Classifier is ", accuracy_score(y_test, y_pred5))
print("The test accuracy score of KNN Classifier is ", accuracy_score(y_test, y_pred6))
```

# We see that Logistic Regression and SVC give the best results.
