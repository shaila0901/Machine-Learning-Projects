Encoding and Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
cat_cols = ['sex','exng','caa','cp','fbs','restecg','slp','thall']
con_cols = ["age","trtbps","chol","thalachh","oldpeak"]

# encoding the categorical columns
x1 = pd.get_dummies(x, columns = cat_cols, drop_first = True)


# instantiating the scaler
scaler = StandardScaler()

# scaling the continuous featuree
x1[con_cols] = scaler.fit_transform(x1[con_cols])
x1.head()
```

# Creating the models

Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x1,y, test_size = 0.2, random_state = 0)
```

Importing the models

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

Logistic Regression

```python
# instantiating the object
logreg = LogisticRegression()

# fitting the object
logreg.fit(X_train, y_train)

y_pred1 = logreg.predict(X_test)
```

SVM

```python
# instantiating the object and fitting
clf = SVC(kernel='linear', C=1, random_state=4).fit(X_train,y_train)

# predicting the values
y_pred2 = clf.predict(X_test)
```

Decision tree

```python
# instantiating the object
dt = DecisionTreeClassifier(random_state = 1)

# fitting the model
dt.fit(X_train, y_train)

# calculating the predictions
y_pred3 = dt.predict(X_test)
```

Random Forest

```python
# instantiating the object
rf = RandomForestClassifier()

# fitting the model
rf.fit(X_train, y_train)

# calculating the predictions
y_pred4 = dt.predict(X_test)
```

Gradient boost

```python
# instantiate the classifier
gbt = GradientBoostingClassifier(n_estimators = 300,max_depth=9,subsample=0.8,max_features=0.2,random_state=2)

# fitting the model
gbt.fit(X_train,y_train)

# predicting values
y_pred5 = gbt.predict(X_test)
```

Checking accuracies

```python
from sklearn.metrics import accuracy_score
```

```python
print("The test accuracy score of Logistic Regression is ", accuracy_score(y_test, y_pred1))
print("The test accuracy score of SVC is ", accuracy_score(y_test, y_pred2))
print("The test accuracy score of Decision Tree Classifier is ", accuracy_score(y_test, y_pred3))
print("The test accuracy score of Random Forest Classifier is ", accuracy_score(y_test, y_pred4))
print("The test accuracy score of Gradient Boosting Classifier is ", accuracy_score(y_test, y_pred5))
```

# Thus we see that Logistic Regression gives us the most accurate model.
